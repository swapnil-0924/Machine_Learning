## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

## Step 2. Import the dataset

```python
data=pd.read_csv('iphones.csv')
data
```

## Step 3. Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

## Step 4. Create LinearRegression Model

```python
lr=LinearRegression()
```

## Step 5. Splitting dataset in to input and output

```python
x=data['iphonenumber'].values.reshape(-1,1)
y=data['price']
```

## Step 6. Fit the Model (Training)

```python
lr.fit(x,y)
```

## Step 7. Prdict the value

```python
lr.predict([[14]]) #price for next iphone
predicted=lr.predict(x)
```

## Step 8. Result Visualization

```python
plt.plot(data['iphonenumber'],data['price'],c='r',marker="*")
plt.plot(data['iphonenumber'],predicted,c='b',marker="^")
```
